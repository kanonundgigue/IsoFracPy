```python
import pathlib, sys
import numpy as np
```

```python
iso_model_dir = pathlib.Path(f"{path_for_repo}/scripts/IsoFracPy").resolve()
sys.path.append(str(iso_model_dir))

from IsotopeFractionationModel import (
    plot_effective_humidity,
    get_init_conditions,
    plot_init_conditions,
    effective_humidity,
    get_init_conditions
)
```

```python
def main():
    temp_sea_list = [  0,   0,   0,  0, 0, 5, 10, 15] # Initial sea surface tempratures
    temp_air_list = [-20, -15, -10, -5, 0, 5, 10, 15]  # Initial air temperatures
    h_air = 1
    surface_wind = 6.5
    plot_effective_humidity(
    temp_air_list, 
    temp_sea_list,
    h_air = h_air
    )

    init_dict = get_init_conditions(
    temp_sea_list,
    temp_air_list,
    h_air = h_air,
    surface_wind = surface_wind,
    )
    plot_init_conditions(init_dict)
```

```python
main()
```
